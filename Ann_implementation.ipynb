```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/content/aps_data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.replace("na",np.nan)
```

```python
df.isnull().sum()
```

```python
pd.options.display.max_rows=1000
```

```python
df.isnull().sum().sum()
```

```python
pd.DataFrame(df.isnull().sum().sort_values(ascending=False)).rename(columns={0:"Null Value Count"})
```

```python
### Filling null values in feature with median of that feature
for feature in [feature for feature in df.columns if feature not in ['class']]:
  df[feature] = df[feature].fillna(df[feature].median())
```

```python
df.isnull().sum().sum()
```

```python
pd.DataFrame(df.isnull().sum().sort_values(ascending = False)).rename(columns={0:"Null Value Count"})
```

```python
df.head()
```

```python
X = df.drop("class",axis=1)
```

```python
y = df["class"]
```

```python
X.info(verbose=True, show_counts=True)
```

```python
y = pd.get_dummies(y,drop_first=True)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Import keras libraries and packages
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
classifier = Sequential()

classifier.add(Dense(units = 10, kernel_initializer='he_uniform',activation='relu',input_dim=170))

classifier.add(Dense(units = 10, kernel_initializer='he_uniform',activation='relu'))

classifier.add(Dense(units = 1, kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
classifier.compile(optimizer ='adam', loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
X_train
```

```python
model_history = classifier.fit(X_train,y_train,batch_size=10,epochs=30, validation_split=0.25)
```

```python
model_history.history.keys()
```

```python
model_history.history['accuracy']
```

```python
# list all data in history

print(model_history.history.keys())
# summarize history for accuracy
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title("model_accuracy")
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['Train','Test'], loc='upper left')
plt.show()
```

```python
X_test.shape
```

```python
y_pred = classifier.predict(X_test)
```

```python
#for the sigmoid my threshold is 0.5
y_pred = y_pred>0.25
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
import time
import os

def savemodel_path(model_dir = "SAVED_MODELS"):
  os.makedirs(model_dir,exist_ok=True)
  fileName = time.strftime("Model_%Y_%m_%d_%H_%M_%S.h5")
  model_path = os.path.join(model_dir,fileName)
  print(f"Your model will be save at the following location\n{model_path}")
  return model_path
```

```python
classifier.save(savemodel_path())
```
